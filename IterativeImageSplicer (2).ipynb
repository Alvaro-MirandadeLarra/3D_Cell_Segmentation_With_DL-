```python
import os
import cv2
import numpy as np

# Define the input and output directories
input_dir = 'raw_data'
output_dir = 'strato3D_raw'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get a list of all .tif files in the input directory
tif_files = [f for f in os.listdir(input_dir) if f.endswith('.tif')]

# Iterate over each 3D image
for tif_file in tif_files:
    # Load the 3D image
    image = cv2.imread(os.path.join(input_dir, tif_file), cv2.IMREAD_UNCHANGED)

    # Generate 2D images along XY, XZ, and YZ coordinates
    xy_image = image[:, :, 0]  # XY plane
    xz_image = image[:, 0, :]  # XZ plane
    yz_image = image[0, :, :]  # YZ plane

    # Save the 2D images with appropriate names
    base_name = os.path.splitext(tif_file)[0]
    cv2.imwrite(os.path.join(output_dir, f'{base_name}_XY.tif'), xy_image)
    cv2.imwrite(os.path.join(output_dir, f'{base_name}_XZ.tif'), xz_image)
    cv2.imwrite(os.path.join(output_dir, f'{base_name}_YZ.tif'), yz_image)

print("2D images generated and saved successfully.")
```
